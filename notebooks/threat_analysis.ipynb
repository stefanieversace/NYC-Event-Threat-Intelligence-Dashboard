# NYC Event Threat Intelligence Analysis

This notebook explores simulated threat intelligence data related to events, venues, and public gatherings in New York City.

## Objectives
- Identify patterns in incidents
- Analyse risk distribution
- Understand operational impact

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
sns.set_theme()

df = pd.read_csv("../data/sample_events.csv")
df.head()
```

```python
df.info()
df.describe()
```

## Risk Distribution

```python
df["risk"].value_counts().plot(kind="bar")
plt.title("Distribution of Risk Levels")
plt.show()
```

## Geographic Analysis

```python
df["borough"].value_counts().plot(kind="bar")
plt.title("Incidents by Borough")
plt.show()
```

## Priority vs Risk

```python
sns.boxplot(x="risk", y="priority_score", data=df)
plt.title("Priority Score by Risk Level")
plt.show()
```

## Time-Based Analysis

```python
df["published_at"] = pd.to_datetime(df["published_at"])
df["hour"] = df["published_at"].dt.hour

df.groupby("hour").size().plot()
plt.title("Incident Frequency by Hour")
plt.show()
```

## Asset Impact Analysis

```python
df["protected_assets"].value_counts().plot(kind="bar")
plt.title("Affected Asset Types")
plt.show()
```

## High-Risk Case Study

```python
high_risk = df[df["risk"] == "HIGH"]
high_risk[["alert_id", "title", "location", "impact"]]
```

## Simulated Response Model

```python
def simulate_response(row):
    if row["risk"] == "HIGH":
        return "Immediate escalation"
    elif row["risk"] == "MEDIUM":
        return "Enhanced monitoring"
    else:
        return "Routine monitoring"

df["response"] = df.apply(simulate_response, axis=1)

df[["alert_id", "risk", "response"]]
```

## Key Findings

- High-risk incidents cluster around major venues
- Transport disruptions impact event operations
- Crowd safety is the most critical concern

## Recommendations

- Increase monitoring at key venues
- Integrate transport intelligence
- Improve crowd management strategies
